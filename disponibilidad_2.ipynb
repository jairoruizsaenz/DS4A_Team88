```python
def melt_df(df):
    #Hours
#     hours= [i for i in range(0,24)]
    hours = list(range(0, 24))
    
    #Convert hours from integers to string, column names
    hours_str = list(map(str, hours))
            
    #Melt original dataframe
    df_melt = pd.melt(df,
                      id_vars=['Fecha', 'Recurso', 'Código Recurso', 'Tipo Generación','Código Agente', 'Archivo'],
                      value_vars=hours_str,
                      var_name='Hour',
                      value_name='Disponibilidad')
    
    #Convert string to float
    df_melt["Hour"] = pd.to_numeric(df_melt["Hour"])
    
    #Add hours to the date which only had considered the day
    df_melt['Fecha'] = pd.to_datetime(df_melt['Fecha'], format="%Y-%m-%d")
    df_melt["Fecha"] = df_melt["Fecha"] + pd.to_timedelta(df_melt["Hour"],unit='H')
    
    #Order de data
    df_melt.sort_values(by=['Fecha'], inplace=True, ascending=True)
    
    #Reset index
    df_melt=df_melt.reset_index(drop=True).drop('Hour', axis = 1).set_index('Fecha')
    
    df_melt['Date'] = df_melt.index.date
    df_melt['Year'] = df_melt.index.year
    df_melt['Month'] = df_melt.index.month
    df_melt['Day'] = df_melt.index.day
    df_melt['Weekday'] = df_melt.index.weekday # Monday=0
    df_melt['Hour'] = df_melt.index.hour

    return df_melt
```

```python
df_raw = melt_df(data)
df_raw
```

```python
list(df_raw)
```

```python
def custom_label(year, month):
    return str(year) + '-' + str(month).rjust(2, '0')
        
df_raw['Year-Month'] = df_raw.apply(lambda x: custom_label(x['Year'],x['Month']),axis=1)
```

```python
df=df_raw[df_raw['Year']<2021]
```

```python
data_termica=df[df["Tipo Generación"]=="TERMICA"].groupby(["Year"])["Disponibilidad"].sum().reset_index()
data_termica["YoY"]=(((data_termica["Disponibilidad"]/data_termica["Disponibilidad"].shift(1))-1)*100)
data_termica["Tipo Generación"]="Térmica"
data_termica.head(3)
```

```python
data_hidraulica=df[df["Tipo Generación"]=="HIDRAULICA"].groupby(["Year"])["Disponibilidad"].sum().reset_index()
data_hidraulica["YoY"]=(((data_hidraulica["Disponibilidad"]/data_hidraulica["Disponibilidad"].shift(1))-1)*100)
data_hidraulica["Tipo Generación"]="Hidráulica"
data_hidraulica.head(3)
```

```python
data_solar=df[df["Tipo Generación"]=="SOLAR"].groupby(["Year"])["Disponibilidad"].sum().reset_index()
data_solar["YoY"]=(((data_solar["Disponibilidad"]/data_solar["Disponibilidad"].shift(1))-1)*100)
data_solar["Tipo Generación"]="Solar"
data_solar.head(3)
```

```python
data_total=df.groupby(["Year"])["Disponibilidad"].sum().reset_index()
data_total["YoY"]=(((data_total["Disponibilidad"]/data_total["Disponibilidad"].shift(1))-1)*100)
data_total["Tipo Generación"]="Total"
data_total.head(3)
```
