```python
import matplotlib.pyplot as plt
import numpy as np


labels = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
# d1 = [20, 34, 30, 35, 27]
# d2 = [25, 32, 34, 20, 25]
# d3 = [25, 32, 34, 20, 25]

d1 = hourly_production1[hourly_production1['Tipo Generación'] == 'HIDRAULICA'].drop(columns=['mes', 'Tipo Generación', 'Generación Total',
 'dia']).reset_index(drop=True).sum(axis=1).values

d2 = hourly_production1[hourly_production1['Tipo Generación'] == 'SOLAR'].drop(columns=['mes', 'Tipo Generación', 'Generación Total',
 'dia']).reset_index(drop=True).sum(axis=1).values

d3 = hourly_production1[hourly_production1['Tipo Generación'] == 'TERMICA'].drop(columns=['mes', 'Tipo Generación', 'Generación Total',
 'dia']).reset_index(drop=True).sum(axis=1).values


x = np.arange(len(labels))  # the label locations
width = 0.2  # the width of the bars

fig, ax = plt.subplots(figsize=(15,6))
rects1 = ax.bar(x - width, d1, width, label='HYDRAULIC')
rects2 = ax.bar(x , d2, width, label='SOLAR')
rects3 = ax.bar(x + width, d3, width, label='TERMAL')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Power (KW/h)')
ax.set_title('Monthly Power generation by source')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# ax.bar_label(rects1, padding=3)
# ax.bar_label(rects2, padding=3)
# ax.bar_label(rects3, padding=3)

fig.tight_layout()
plt.show()
```

```python
df1 = pd.read_excel("Consumo_De_Combustible_(MBTU)_2020.xlsx", header=2)
df1.head()
```

```python
Chartfuel=TotalCombustible.reset_index()

Chartfuel.plot.bar(x="Combustible", y="Consumo Combustible (MBTU)", figsize=(13,6), title="Consumo Combustible")
#plt.ylim(17000000000,18200000000)b
```

```python
df1["new"]=df1["Fecha"]+df1["Recurso"]+"TERMICA"
df1.head()
```

```python
df2=df1[["new", "Combustible", "Consumo Combustible (MBTU)"]]
df2=df2[df2["Consumo Combustible (MBTU)"]!=0]    
df2.head()
```

```python
df3=df.merge( df2, how="inner", on="new")
df3["Generación Total"]=df3.sum(axis=1)
df3=df3[["Fecha", "Recurso", "Combustible", "Consumo Combustible (MBTU)", "Generación Total" ]]
```

```python
df3comb=df3.groupby(["Combustible"]).sum()
df3comb
```

```python
df3rec=df3.groupby(["Recurso", "Combustible"]).sum()
df3rec=df3rec.sort_values('Generación Total',ascending=False)
df3rec.head(10)
```

```python
Chartfuelgeneration=df3rec.reset_index()

Chartfuelgeneration.plot.bar(x="Recurso", y=[ "Generación Total"], figsize=(13,6), title="Generación Total", color="orange")
Chartfuelgeneration.plot.bar(x="Recurso", y=["Consumo Combustible (MBTU)"], figsize=(13,6), title="Consumo Combustible")

plt.legend(loc='upper left')
#plt.ylim(17000000000,18200000000)b
```

```python
df3comb["ratio"]=df3comb["Generación Total"]/df3comb["Consumo Combustible (MBTU)"]
df3comb=df3comb.sort_values("Generación Total",ascending=False)
df3comb
```

```python
Chartfuelproduction=df3comb.reset_index()

Chartfuelproduction.plot.bar(x="Combustible", y="Generación Total", figsize=(13,6), title="Consumo Combustible")
#plt.ylim(17000000000,18200000000)b
```
